```python
import pandas as pd
import numpy as np
cars = pd.read_csv("https://gist.githubusercontent.com/noamross/e5d3e859aa0c794be10b/raw/b999fb4425b54c63cab088c0ce2c0d6ce961a563/cars.csv")
cars.head()
```

```python
cars.describe()
```

```python
cars = cars.drop(['Unnamed: 0'], axis=1)
```

```python
cars_corr=cars.corr()
import seaborn as sns
sns.heatmap(cars_corr,annot=True)
```

```python
x=np.array(cars['disp']).reshape(-1,1)
y=np.array(cars['cyl']).reshape(-1,1)

x_train = x[:-20]
y_train = y[:-20]

x_test = x[-20:]
y_test = y[-20:]

from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score,mean_squared_error
model = LinearRegression()
model.fit(x_train,y_train)
y_pred = model.predict(x_test)

import matplotlib.pyplot as plt
plt.scatter(x_train,y_train,color="black")
plt.plot(x_test,y_pred,color='red')
plt.show()
```

```python
r2 = r2_score(y_test,y_pred)
r2
# this is showing probability that your anstwer is 0.79*100 % correct
```

```python
# its +-0.74 in our answer
# Its know as error
# ex ans is 100 we can show it like 100+0.79 or 100-0.79   (0.79 is mean_sqr_error)
mean_squared_error(y_test,y_pred)
```
