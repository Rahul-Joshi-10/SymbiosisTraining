OUTLIRES:

* suppose we have 100 points between 60 to 90
* Q1 = 25 points from 60 to 68 
* Q2 = 25 points from 68 to 74

* IQR = Inner Quartile Range = ( Q3 - Q1 )

* To find out outlires which are too much higher we have formula  (+ 1.5 * IQR from Q3)
* MAX_Outlires = Q3 + 1.5 * IQR
* To findout outlires which are too much lower (smaller )we have formula (- 1.5 * IQR from Q1)
* MIN_Outlires = Q1 - 1.5 * IQR


```python
import matplotlib.pyplot as plt
import pandas as pd 
import numpy as np
```

```python
# Example using list
list1 = [22,25,25,25,26,26,27,29,30,31,31,31,102]
data=pd.DataFrame(list1)
size = len(list1)
Q1 = round(size * 0.25)
Q1 = list1[Q1]    # return element at 25%
Q3 =  round(size * 0.75) 
Q3 = list1[Q3]    # return element at 75%

# Inner Quartile Range
IQR = Q3 - Q1

Max_outlire = Q3 + 1.5 * IQR
Min_outlire = Q1 - 1.5 * IQR

outlire = []
new_list = []
for i in list1:
    if i > Max_outlire or i<Min_outlire:
        outlire.append(i)
    else:
        new_list.append(i)

outlire,new_list

```

```python
dataset = pd.read_csv("https://raw.githubusercontent.com/selva86/datasets/master/BostonHousing.csv")
dataset.info()
```

```python
Q1 = np.percentile(column1,25) # this will return element at 25% 
Q3 = np.percentile(column1,75) # this will return element at 75% 

IQR = Q3-Q1

Max_outlire = Q3 + 1.5 * IQR
Min_outlire = Q1 - 1.5 * IQR

outlires = []
non_outlires = []

for i in column1:
    if i > Max_outlire or i< Min_outlire:
        outlires.append(i)
    else:
        non_outlires.append(i)

print(non_outlires)
print(outlires)

plt.boxplot(non_outlires)
```

L1 Normalization:
* each data is devided by sumof(absolute of data)
* XL1 = X/SUM(abs(X))
* thus this teq of normalization is also know as least absolute Deviation or manhatan Norm
* Where to use L1 Normalizaion:
     * L1 Normalization can be use with data with many 0's or sparcly distributed data
     * Text and image processing 

Advantages:
* L1 Normalization is less sensitive to outlires
* Better than L2 Normalization  

L2 Normalizaion:
* This method scales feature vector to length of 1
* XL2 = X/sqrt(SUM(X²))
