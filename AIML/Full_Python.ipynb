```python
# Map() -> fucntion which take 2 argument
# 1st para -> function
# 2nd para -> iterator like list or array so something

list1=[1,2,3,4,5,6,7,8]

def square(n):
    return n*n
result = map(square,list1)
final_result = list(result)  # make object_of_map(result) to list

print(final_result)
```

```python
list1=[1,2,3,4,5,6,7,8]

def square(n):
    return n*n
result = map(square,list1)
final_result = iter(result)  # make object_of_map(result) to list

for i in list1:
    print(next(final_result))
```

```python
class Student:
    def __init__(self,name,roll,marks):
        self.name=name
        self.marks=marks
        self.roll=roll

    def isPass(self):
        if self.marks<=40:
            return 40-self.marks
        else:
            return 0

ajinkya = Student('ajinkya',12,20)
print(f"{ajinkya.name} got grace of : {ajinkya.isPass()}")
```
