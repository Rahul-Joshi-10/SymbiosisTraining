* Standardize feature by removing the mean and scaling to unit variance the standard score od a sample x is calculated
* Because
* z = (x-u)/s where z is score u is mean and x is the number s is std
* where u is mean of training samples or zero with_mean=False and s is standard deviation of training samples or zero with_std=False

```python
data = [[0,0],[0,0],[1,1],[1,1]] # making 2 columns
scaler = StandardScaler()
print(scaler.fit(data))
StandardScaler()
print(scaler.mean_)
print(scaler.transform(data))
```
