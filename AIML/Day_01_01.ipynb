# Before Using LinearRegresion We are supposed to check Co relation between 2 Variables

```python
import pandas as pd
import numpy as np
df = pd.read_csv('/content/sample_data/california_housing_test.csv')
```

# - is showing negative Sensitivity or correlation

## Observation = Rows
## Features = coloumns
## Arbitary data = created by user ex: [1,2,3,4,5,6,7,8]


## This model only accept numpy objects or arrays
## It will not accept DataFrames of Pandas
## we can create object of LinearRegression() manually or Using (AS) shown above
## model = LinearRegression()
## positional arguments first one always independant other always dependant
## model.fit(x,y)

```python
x = np.array([5,15,25,35,45,55]).reshape(-1,1)
y = np.array([5,20,14,32,22,38])
```

```python

from sklearn.linear_model import LinearRegression
model = LinearRegression()
model.fit(x,y)
```

## Range of score is 0 to 1
## dots on line in %
## 0 = dots more under line
## 1 = dots more upside line
## How much correct is your answere ( probability )

```python
model.score(x,y)
```

## This will show the interception of line to point
## Straight line intercepts AT point Y that value is shown By
## In short starting point of straight Line

```python
model.intercept_
```

## This is known as slope or sensitivness

```python
model.coef_
```
