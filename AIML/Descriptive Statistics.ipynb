* Nominal: Unordered categorical variables
* Ordinal: There is an ordering but no implication of equal
distance between the different points of the scale.
* Interval: There are equal differences between successive
points on the scale but the position of zero is arbitrary.
* Ratio: The relative magnitudes of scores and the differences
between them matter. The position of zero is fixed.

```python
# Inferential -> using or selected data point to study or predict dataset
import scipy
mean = 0
var = 2
sigma = np.sqrt(var)   # std dev
x  = np.linspace(mean-3*var,mean+3*var,100)
_ = plt.plot(x,scipy.stats.norm.pdf(x,mean,sigma))
```

```python
# Probability desity function
fvalues = np.linspace(-1,5,100)
plt.plot(fvalues, scipy.stats.f.pdf(fvalues,1,30),'b-',label="F(1,30)")
plt.plot(fvalues, scipy.stats.f.pdf(fvalues,5,30),'r-',label="F(5,30)")
plt.legend()
```
