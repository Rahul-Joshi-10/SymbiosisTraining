```python
from sklearn import datasets, linear_model
from sklearn.metrics import mean_squared_error, r2_score
# r2 is 0 to 1 coz its probability
import numpy as np

# Load Diabetes Dataset
# in x we are loading all coloum and in y we are only loading one coloum
diabetes_X, diabetes_y = datasets.load_diabetes(return_X_y=True)
print(diabetes_X) # this one having all columns
# now x have 10 colum and y have only one column
# remove 9 coloum and only select one by index
diabetes_X = diabetes_X[:,np.newaxis,2]
diabetes_X # now this one have only one column of index 2
```

```python
# now devide x and y into 4 group
# one for testing and one for training
# ratio will be 80:20 80% for training and 20% for testing [Matrix] -> this is random data if this datafeeded to moddel then how will it behave that is done
diabetes_X_train = diabetes_X[:-20] # 80% data for tain
diabetes_X_test  = diabetes_X[-20:]  # 20% data for test
diabetes_y_train = diabetes_y[:-20] # 80% data for tain
diabetes_y_test  = diabetes_y[-20:]  # 20% data for test
```

```python
# Fitting 80% data to model
from sklearn.linear_model import LinearRegression
model = LinearRegression()
model.fit(diabetes_X_train,diabetes_y_train)
```

```python
y_predict = model.predict(diabetes_X_test)
```

```python
import matplotlib.pyplot as plt
plt.scatter(diabetes_X_test,diabetes_y_test,color="black")
plt.plot(diabetes_X_test,y_predict,color="red")
plt.show()
```
