```python
# write a program to implement stack in python push and pop

class Stack:

    def __init__(self,size):
        self.top = -1
        self.stack = [None]*size
        self.size = size
        print(self.stack)

    def push(self, item):
        if self.top == self.size:
            print("Stack is FUll")
        else:
            self.top += 1
            self.stack[self.top] = item

    def pop(self):
        if len(self.stack)!=0:
            temp = self.top
            self.top-=1
            return self.stack[temp]
        else:
            print("cant't delete stack is empty")
            return 0

    def showStack(self):
        temp = self.top
        while temp!=-1:
            print(self.stack[temp])
            temp-=1

stack = Stack(10)
stack.push(10)
stack.push(9)
stack.push(8)
stack.push(7)
stack.showStack()
stack.pop()
stack.showStack()
```

```python
# Create linkedlist in python

class Node:
    def __init__(self,data):
        self.data = data
        self.next = None


class linkedlist:
    def __init__(self):
        self.head = None

    def insertAtBegin(self,data):
        new_node = Node(data)
        if self.head == None:
            self.head = new_node
            return
        new_node.next = self.head
        self.head = new_node

    def insertAtEnd(self,data):
        new_node = Node(data)
        if self.head is None:
            self.head = new_node
            return

        current_node = self.head
        while(current_node.next):
            current_node = current_node.next

        current_node.next = new_node

    def removeFristNode(self):
        if self.head == None:
            return
        self.head = self.head.next

    def removeLastNode(self):
        if self.head == None:
            return
        current_node = self.head
        while current_node.next != None and current_node.next.next != None:
            current_node = current_node.next

        current_node.next = None

    def printAll(self):
        current_node = self.head
        while current_node:
            print(current_node.data, end="->")
            current_node = current_node.next
        print()

    def removeNode(self, data, index):
        if index == 0:
            self.insertAtBegin(data)

        position = 0
        current_node = self.head
        while current_node != None and position +1 != index:
            position += 1
            current_node = current_node.next

        if current_node != None:
            new_node = Node(data)
            new_node.next = current_node.next
            current_node.next = new_node
        else:
            print("Index is not present...")


    def updateNode(self, data, index):
        current_node = self.head
        position = 0
        if index == position :
            current_node.data = data 

        else:
            while current_node != None and position+1 != index:
                position += 1
                current_node = current_node.next

            if current_node != None:
                current_node.data = data

            else:
                print("Index Dose not exist") 



link = linkedlist()

link.insertAtBegin(10)
link.insertAtEnd(20)
link.insertAtEnd(30)
link.insertAtEnd(40)
link.printAll()
link.updateNode(25,2)
link.printAll()
link.removeFristNode()
link.printAll()
link.removeLastNode()
link.printAll()
            
```

#### Dynamic Programing:
* similar to devide and conquer approch dynamic programming also combines the solution to sub problem it mainly used where soloution of subproblem is needed repetadly the computed solutions are stored into table, so that we don't have to recompute i.e overlapping sub problem exist

* Examples: Floyd-Warshall and Bellman-Ford are typicall exmples of DP

Steps of Dp:
* Characterize the structure of Optimal soloution (Sorted Manner)
* Recursively define the value of optimal soloution
* compute values into recursive manner (bottom up approch)
* Construct an optimal soloution form the computed information

Devide and Conquer : problem devided into Disjoint set of sub problems solved recursively and then combined them to solve orignal problem

DP: it used when sub problem are interdependant [over lapping of subproblems can happen] (sometimes they share same sub problem) in short problem is solved once and store into table

Element of Dynamic Programming
* substructure : devide problem into subproblem
* table structure : store the result into table
* bottom up approch : start form bottom 

Advantages of Dynamic Programming (DP):
* Avoids recomputing the same subproblems multiple times, leading to significant time savings.
* Ensures that the optimal solution is found by considering all possible combinations.
* Breaks down complex problems into smaller, more manageable subproblems.

Applications of DP:
* Shortest path
* Longest common Subsequence
* Maximize CPU Cycles (no cycle get wested)
